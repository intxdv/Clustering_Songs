```python
data = data.dropna()
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
data[['explicit','mode']] = data[['explicit','mode']].astype(int)
```

```python
num_cols = ['popularity','duration_ms','danceability','energy'
            ,'loudness','speechiness','acousticness','instrumentalness'
            ,'valence','tempo', 'liveness']
cat_cols = ['track_name','artists','album_name','track_genre'
            ,'time_signature', 'explicit', 'mode',  'key']
```

```python
# Checking for duplicate records
print("Number of duplicate records: ", len(data[data.duplicated()]))

# Removing duplicate records
data = data[~data.duplicated()]
```

```python
# Checking if 1 track has multiple records, with different values for attributes like popularity
track_counts = data['track_id'].value_counts()
multiple_tracks = track_counts[track_counts > 1]

if not multiple_tracks.empty:
    print("Tracks with multiple rows:")
    print(multiple_tracks)
else:
    print("No tracks have multiple rows.")
```

```python
# Data cleaning to get 1 record per track identified by track_id

# Custom function to join mode values as a comma-separated string
def mode_as_string(series):
    unique_modes = series.mode().astype(str).unique()  # Get unique mode values
    return ', '.join(unique_modes)

# Group by 'track_id' and aggregate
aggregated = data.groupby(['track_id']).agg({
    'track_name': pd.Series.mode,
    'artists': mode_as_string,
    'album_name': pd.Series.mode,
    'explicit': 'max',
    'track_genre': mode_as_string,
    'time_signature': 'max',
    'key': 'max',
    'mode': 'max',
    'popularity': 'mean',
    'duration_ms': 'mean',
    'danceability': 'mean',
    'energy': 'mean',
    'loudness': 'mean',
    'speechiness': 'mean',
    'acousticness': 'mean',
    'instrumentalness': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'liveness': 'mean'
}).reset_index()

# Display the aggregated result
data_cleaned = aggregated.copy()
```

```python
def map_genre_groups(x):
    if ',' in x:
        x = x.split(",")
    else:
        x = [x]
    
    genre_map = {
        'pop': 'pop',
        'techno': 'electronic',
        'electro': 'electronic',
        'dubstep': 'electronic',
        'house': 'electronic',
        'soul': 'soul',
        'funk': 'soul',
        'bluegrass': 'country',
        'rock': 'rock',
        'metal': 'metal',
        'grunge': 'rock',
        'jazz': 'jazz',
        'samba': 'dance',
        'pagode': 'dance',
        'tango': 'dance',
        'dance': 'dance',
        'blues': 'blues',
        'reggae': 'reggae',
        'ska': 'reggae',
        'r-n-b': 'rnb',
        'classical': 'classical',
        'opera': 'classical',
        'folk': 'folk',
        'hardcore': 'electronic',
        'hardstyle': 'electronic',
        'trance': 'electronic'
    }
    
    genre_group = []
        
    for genre in x:
        mapped=False
        for genre_key in genre_map.keys():
            if genre_key in genre:
                genre_group.append(genre_map[genre_key])
                mapped=True
                break
        if mapped==False:
            genre_group.append('other')
    
    genre_group =list(set(genre_group))
            
    
    return ",".join(genre_group)

```

```python
# Creating new genre category column
data_cleaned['genre_category'] = data_cleaned['track_genre'].apply(lambda x: map_genre_groups(str(x).lower()))
data_cleaned.head()
```

# Clustering Songs

Songs can be grouped into clusters based on different numerical and categorical features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from category_encoders import TargetEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
```

```python
# Picking final set of numerical and categorical features
# Dropping 'loudness' as a numerical feature since it has high correlation with multiple features

cluster_num_cols = ['danceability','energy'
            ,'speechiness','acousticness','instrumentalness', 'liveness']
cluster_cat_cols = ['mode', 'key', 'time_signature'] #['genre_category']

cluster_data = data_cleaned.copy()
cluster_data.shape
```

```python
cluster_data.head()
```

```python
# Build a Pipeline

# Create a preprocessor with StandardScaler and OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), cluster_num_cols),
        ('cat', OneHotEncoder(), cluster_cat_cols),
    ],
    remainder='drop'
)
```

```python
def get_kmeans_cluster_size(X):
    # Step 2: Calculate silhouette scores for different values of k
    silhouette_scores = []
    k_values = range(2, 11)  # Testing for k from 2 to 10

    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        clusters = kmeans.fit_predict(X)
        score = silhouette_score(X, clusters)
        silhouette_scores.append(score)

    # Step 3: Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, silhouette_scores, marker='o')
    plt.title('Silhouette Score for Optimal k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.xticks(k_values)
    plt.grid()
    plt.show()

    # Display the optimal k based on the highest silhouette score
    optimal_k = k_values[np.argmax(silhouette_scores)]
    print(f'Optimal number of clusters: {optimal_k}')   

def plot_clusters(X,labels):
    pca = PCA(n_components=2, svd_solver='randomized')
    X_reduced = pca.fit_transform(X)
    
    # Plot results
    plt.figure(figsize=(8, 6))
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis', marker='o')
    plt.title('KMeans Clustering Results')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar()
    plt.show()
```

```python
# Apply the preprocessing for cleaned data
# Setting common datatypes for all numerical features and categorical features respectively
for col in cluster_num_cols:
    cluster_data[col] = cluster_data[col].astype('float')

for col in cluster_cat_cols:
    cluster_data[col] = cluster_data[col].astype('str')

X = preprocessor.fit_transform(cluster_data)
X.shape
```

```python
get_kmeans_cluster_size(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
cluster_data['Cluster'] = labels
X.shape, labels.shape
```

```python
plot_clusters(X,labels)
```
